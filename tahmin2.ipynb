```python
# Ensure googletrans-temp is installed and in use

from deep_translator import GoogleTranslator
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
data = pd.read_csv("C:/Users/Dell/Desktop/DS/DSnot/akp_tahmin/data.csv")
df = pd.DataFrame(data)

# Load data

def temizle_metni(text):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    return None

# Apply cleaning
df["temizText"] = df["text"].apply(temizle_metni)

# Translate and filter text
translated_texts = []
for i in df["temizText"]:
    if i:
        try:
            translated_text = GoogleTranslator(source='auto', target='en').translate(i)
            translated_texts.append(translated_text)
        except Exception as e:
            print(f"Error translating text '{i}': {e}")
            translated_texts.append(None)
    else:
        translated_texts.append(None)

# Remove stopwords
filtered_translations = []
for translated_text in translated_texts:
    if translated_text:
        filtered_translation = ' '.join([word for word in translated_text.split() if word.lower() not in stop_words])
        filtered_translations.append(filtered_translation)
    else:
        filtered_translations.append(None)

df["filteredText"] = filtered_translations
df = df.dropna(subset=['filteredText'])

print(df[['text', 'temizText', 'filteredText']])

```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Verilerinizi eğitim ve test olarak ayırın
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model tanımlamaları
models = {
    "Gaussian Naive Bayes": GaussianNB(),
}

# Modelleri eğitip doğruluklarını karşılaştırın
accuracies = {}

for model_name, model in models.items():
    model.fit(X_train.toarray(), np.ravel(y_train))  # Modeli eğitin
    y_pred = model.predict(X_test.toarray())          # Tahmin yapın
    accuracy = accuracy_score(y_test, y_pred)         # Doğruluğu hesaplayın
    accuracies[model_name] = accuracy                  # Sonuçları saklayın

# Sonuçları yazdırın
for model_name, accuracy in accuracies.items():
    print(f"{model_name} Modelinin Doğruluğu: {accuracy:.2f}")
```

çözüm Yolu:

Veri Ön İşleme:

Temizleme: Cümlelerdeki noktalama işaretleri, sayılar ve gereksiz karakterler gibi gürültüyü temizleyin.
Tokenleştirme: Cümleleri kelime veya alt kelime birimlerine (token) ayırın.
Küçük Harfe Çevirme: Tüm kelimeleri küçük harfe çevirerek tutarlılık sağlayın.
Durulama: Sık kullanılan ancak anlamsal olarak çok katkı sağlamayan kelimeleri (stop words) kaldırın.
Köksöz Bulma: Kelimeleri köksözlerine indirgeyerek kelime varyasyonlarını tek bir köksöz altında toplayın (stemming veya lemmatization).
Vektörleştirme:

Bag-of-Words: Kelimelerin metin içindeki sıklıklarını sayarak sayısal bir vektör oluşturun.
TF-IDF: Kelimelerin bir belge içindeki önemini ve tüm belgedeki nadirliğini dikkate alarak ağırlıklandırılmış bir vektör oluşturun.
Word Embeddings: Kelimeleri anlamsal ve sözdizimsel olarak benzer kelimelerle birlikte düşük boyutlu bir uzaya yerleştiren yoğun vektörler kullanın (Word2Vec, GloVe, BERT gibi).


Bu tür bir metin sınıflandırma projesi için aşağıdaki adımları izleyebilirsiniz:

1. Veri Ön İşleme
Öncelikle, verinizi temizlemeniz ve ön işlemeniz gerekiyor. Bu adımlar şunları içerebilir:

Büyük/küçük harf dönüşümü: Metni tek bir formatta (genellikle küçük harf) tutun.
Noktalama işaretleri ve özel karakterler: Gereksiz noktalama işaretlerini ve özel karakterleri kaldırın.
Kelime kökleri: Kelimeleri köklerine indirgemek için lemmatization veya stemming uygulayın.
Stop word'ler: Anlamsız kelimeleri (ve, ve, ama gibi) kaldırın.
2. Özellik Çıkarma
Metinleri sayısal verilere dönüştürmek için çeşitli yöntemler kullanabilirsiniz:

Bag of Words (BoW): Metni kelime frekanslarıyla temsil eder.
TF-IDF (Term Frequency-Inverse Document Frequency): Kelimelerin metin içindeki önemini ölçer.
Word Embeddings: GloVe veya Word2Vec gibi yöntemlerle kelimeleri vektörlere dönüştürerek daha anlamlı temsil sağlar.
3. Model Seçimi
Sınıflandırma algoritmaları arasında seçim yapabilirsiniz. Yaygın olarak kullanılan bazı algoritmalar şunlardır:

Naive Bayes: Metin verileri için basit ve etkili bir yöntem.
Logistic Regression: İkili sınıflandırma için iyi sonuçlar verebilir.
Support Vector Machines (SVM): Yüksek boyutlu veri setlerinde iyi performans gösterir.
Random Forest veya Gradient Boosting: Daha karmaşık modeller, genellikle daha iyi sonuçlar verir.
Derin Öğrenme: RNN veya LSTM gibi modeller, özellikle büyük veri setlerinde etkili olabilir.
4. Modeli Eğitme
Verilerinizi eğitim ve test setlerine ayırın (örneğin, %80 eğitim, %20 test). Seçtiğiniz modelle eğitim setinde eğitin ve ardından test setinde doğruluğunu ölçün.

5. Değerlendirme
Modelinizin performansını değerlendirmek için aşağıdaki metrikleri kullanabilirsiniz:

Doğruluk: Doğru sınıflandırılan örneklerin oranı.
F1 Skoru: Kesinlik ve hatırlamanın birleşimi.
Confusion Matrix: Sınıflandırma hatalarını görselleştirmek için kullanılır.
6. İyileştirme
Modelinizin performansını artırmak için şu yöntemleri düşünebilirsiniz:

Hyperparameter Tuning: Modelin parametrelerini optimize edin.
Özellik Seçimi: Sadece en iyi özellikleri kullanarak modelinizi basitleştirin.
Ensemble Yöntemleri: Farklı modelleri birleştirere
